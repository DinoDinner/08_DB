#### 비모수 검정 과제

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
```

```python
# 랜덤한 100개 데이터 생성

group1 = np.random.randint(0,70,100)
group2 = np.random.randint(30,100,100)

print(group1)
print(group2)
```

```python
# 분포 시각화

fig, ax = plt.subplots()
ax.hist(group1, bins=10, alpha=0.5, label='group1')
ax.hist(group2, bins=10, alpha=0.5, label='group2')
ax.set_title('Random Number 100')
ax.set_xlabel('Number')
ax.set_ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Willcoxon Signed-Rank Test 시행

stat, p = wilcoxon(group1,group2)

print("Wilcoxon signed-rank test")
print("-------------------------")
print("stat : ",stat)
print("p-value : ",p)
```

```python
# 결과

if p > 0.05:
    print("Fail to reject H0")
else :
    print("Reject H0")
```

#### 해석

- Willcoxon signed-rank test 결과 독립적인 두 그룹의 유사성 정도는 stat(413.5)으로 확인할 수 있다.

- 두 그룹이 유사한 정도는 귀무가설(H0)이 기각 되었기 때문에 차이가 있다는 걸 알 수 있다.
