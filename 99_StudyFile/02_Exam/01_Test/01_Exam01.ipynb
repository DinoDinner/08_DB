```python
# 상관 분석 과제
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
grade = np.tile([1,2,3],100)
math = np.random.normal(70,5,300)
english = np.random.normal(70,6,300)
korean = np.random.normal(70,3,300)
# print(grade, math, english, korean)

# 데이터 프레임 csv 저장
data_score = pd.DataFrame({'Grade':grade,'Math':math,
                           'English':english,'Korean':korean})
data_score.to_csv('../01_Test/grade_score.csv',index=False)

# csv 파일 불러오기
score = pd.read_csv('../01_Test/grade_score.csv')

# 다중 상관 분석 수행
score_corr = score.corr()
print(score_corr)
```

```python
# 히트맵 출력
sns.heatmap(score_corr,cmap="magma",
            annot=True,fmt='.2f',cbar=True)
```

```python
# 데이터 과목별로 나누기
score_math = score['Math']
score_english = score['English']
score_korean = score['Korean']

print(score_math.describe(), 
      score_english.describe(), 
      score_korean.describe())
```

```python
# 통계 시각화

# Math
plt.hist(score_math,bins=10)
plt.xlabel('score')
plt.ylabel('Frequency')
plt.title('Math Score Distribution')
plt.show()

# English
plt.hist(score_english,bins=10)
plt.xlabel('score')
plt.ylabel('Frequency')
plt.title('English Score Distribution')
plt.show()

# korean 
plt.hist(score_korean,bins=10)
plt.xlabel('score')
plt.ylabel('Frequency')
plt.title('Korean Score Distribution')
plt.show()
```
