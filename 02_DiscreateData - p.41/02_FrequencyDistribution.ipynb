#### 빈도수

- 이산형 데이터는 정수로 이루어진 데이터
- 주로 명목형 데이터와 순서형 데이터로 구성된다.
- 빈도수를 통해 전체 데이터의 분포를 파악할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 빈도수는 막대그래프를 활용 시각적으로 보는게 좋음
```

```python
# 데이터 생성
data = np.random.randint(1,7, size=100)
# 1~6 사이에 100개 숫자를 랜덤으로 생성

print(data)

# 빈도수 계산
unique, counts = np.unique(data, return_counts=True)
# unique : 고유한 값들을 출력함.
# return=True : 고유한 값들의 개수를 출력해줌.

print(unique, counts)

# 막대 그래프 시각화
fig, ax = plt.subplots()
ax.bar(unique, counts)

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Frequency Distribution')
plt.show()
```
