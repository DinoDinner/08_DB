#### 공간 데이터

1. 지리정보(지도상의 위치정보)와 관련된 데이터
2. 지도상의경계, 거리, 면적, 위치등을 분석하고 시각화할 수 있다.
3. 예시 : 도로교통량예측, 대기오염예측, 지진발생가능성예측

- Point 형식 : 위도와 경도등의 좌표정보를 가지고있는 데이터
- Line 형식 : 도로나 하천과같이 길이와 방향이있는 데이터
- Ploygon 형식 : 도시나 지역과같이 면적이있는 데이터

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 서울시 행정구역 경계 데이터 읽기
seoul_gdf = gpd.read_file('../99_File/02_Data/05_Seoul/seoul_municipalities_geo_simple.shp')

# 그래프 크기 설정
plt.figure(figsize=(10,10))

# 그래프 타이틀 설정
seoul_gdf.plot()

# 그래프 타이틀 설정
plt.title('Seoul Municipalties')

# x,y 축 라벨 설정
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 그래프 표시
plt.show()
```
