#### 산포도

1. 데이터의 분산정도를 나타내는 그래프
2. X축 : 데이터값 / Y축 : 빈도수
3. 산포도가 작을 수록 데이터 값들의 분포가 집중된다.
4. 분산/표준편차는 데이터가 평균에서 얼마나 멀리 퍼져있는 지를 나타내는 지표이다.
5. 산포도와 분산/표준편차는 정의 관계

계산 가능한 데이터
- 분산(variance)
- 표준편차(standard deviation)
- 범위(range)
- 사분위간범위(interquartile range)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 두 변수 간의 상관관계 생성

np.random.seed(0)   # 랜덤 숫자 생성
x = np.random.normal(loc=0, scale=1, size=100)  # 표준편차 = 1 의 숫자 100개
y1 = x+np.random.normal(loc=0, scale=0.1, size=100) # 표준편차 = 0.1
y2 = x+np.random.normal(loc=0, scale=7, size=100)   # 표준편차 = 7
```

```python
# 산포도 그리기 
# y1(표준편차 = 0.1)
fig, ax = plt.subplots()
ax.scatter(x,y1)  # 1번 subplot 그래프에 sin(x)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')
```

```python
# 산포도 그리기 
# y2(표준편차 = 7)
fig, ax = plt.subplots()
ax.scatter(x,y2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')

plt.show()
```

```python
# 비교 그래프

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(x, y1)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y1')
axes[0].set_title("STD 0.1")

axes[1].scatter(x, y2)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y2')
axes[1].set_title("STD 7")

plt.show()
```

- x,y는 양의 상관관계를 가진다.
- y1보다 표준편차가 큰 y2의 그래프가 산포도가 더 큰 것을 확인할 수 있다.
