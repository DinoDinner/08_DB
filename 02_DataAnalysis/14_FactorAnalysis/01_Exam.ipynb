#### 랜덤 데이터셋을 사용한 인자 분석 실습

```python
import numpy as np
from sklearn.decomposition import FactorAnalysis
```

```python
# 데이터 생성
n_samples = 100 # 샘플
n_featrues = 5 # 변수

# 데이터셋을 랜덤하게 설정
x = np.random.randn(n_samples,n_featrues)
print(x)
```

```python
# 인자 분석 수행
n_componets = 2 # 추출할 인자 수
factorAnalysis = FactorAnalysis(n_components=n_componets, random_state=0) # random_state : 시드
x_factor= factorAnalysis.fit_transform(x)

# 결과 출력
print('원본 데이터셋 크기', x.shape)
print(f'인자 분석 결과 인자 개수 : {n_componets} ')
print()
```
