#### 주성분 분석(PCA)

1. 다차원 데이터의 차원을 축소하면서 데이터의 주요한 패턴을 추출하는 비지도 학습 알고리즘
2. 데이터를 더 낮은 차원으로 표현해서 데이터의 복잡성을 감소시키고, 변수 간의 상관 관계 및 패턴을 파악

- 차원 축소
    - 다차원 데이터의 여려 변수들은 상호 연관성을 가지고 있거나 중복되는 정보를 포함 할 수 있음.
    - 다차원 데이터를 고차원 공간에서 저차원 공간으로 변환하여 데이터의 중요한 특징과 변동성 유지하면서 차원 줄임
    - 결과적으로 데이터의 표현이 더 간결하고 해석하기 쉽게 됨

- 주성분 추출
     - 데이터의 분산을 최대로 보존하는 방향 : 주성분(Principal Component)을 찾음
     - 주성분은 원본 데이터의 변동성을 가장 잘 설명하는 변수들의 선형 조합이다.
     - 1번 주성분 : 데이터의 가장 큰 분산, 그 이후 주성분들은 변동성이 점차 감소함

- 정보 유실 최소화
    - 차원 축소를 통해 데이터의 정보를 최대한 보존하려고 함
    - 원본 데이터의 변동성을 설명하는데 가장 큰 기여를 하는 변수들로 구성, 상대적으로 작은 수의 주성분으로 데이터의 주요 특징 잘 표현
    - 저차원으로 표현해서 정보유실 최소화, 계산 및 최적화 효율 높임

- 데이터의 정규화, 표준화 등 전처리 방법
    1. 2차원의 데이터 셋
    2. 데이터 셋의 분포 확인
    3. 공분산 행렬 계산
    4. 고우값과 고유벡터 계산
    5. 주성분 선택
    6. 데이터 투영
    7. 차원 축소

- 주성분 분석(PCA) : 데이터 전처리

    1. 변수 스케일링 : 크기가 큰 변수에 더 큰 가중치 부여
    2. 변수 상관 관계 : 상관 관계가 큰 변수들이 존재할 경우, 변수간 상관 관계를 줄임
    3. 이상치 처리 : 결과 왜곡을 줄이기 위해 이상치를 탐지하고 처리
    4. 결측치 처리 : 누락된 값을 처리하여 결과 왜곡을 줄임
