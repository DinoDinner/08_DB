#### 00. 시계열 데이터의 특징
<br>

- 데이터 특징
    1. 순서 : 시간 순서에 따라 정렬된 데이터
    2. 시간 의존성 : 시간에 따라 변화하는 경향
    3. 계절성 : 주기적인 계절성 패턴을 가짐
    4. 트렌드 : 장기적인 추세나 동향이 존재할 수 있음
    
<br>

<br>

- 시계열 데이터의 기본가정 : 가정을 이해하고 고려하여 패턴과 동향을 분석, 예측

    1. 정상성 가정 : 통계적 특성이 시간에 따라 일정하게 유지되는 것
    2. 독립성 가정 : 시간 의존성을 가지지만, 이전 시점의 데이터가 현재 시점에 영향을 주는 경우가 많음
    3. 잔차의 정규성 가정 : 모델의 잔차가 정규분포를 따른다는 가정

<br>


<br>

- 자기 상관성과 정상성 개념 - 데이터를 분석하고 예측하는데 다양한 시계열 모델 활용

    1. 자기 상관성 : 관측 값들 사이에 상관관계를 의미
    2. 정상성  : 통계적 특성이 시간에 따라 일정하게 유지되는 것을 의미

<br>


- 자기회귀 모델(AR) 및 이동 평균 모델(MA) 소개

    1. 이동 평균 모델(Moving Average Model)
        - 이동평균 : 이전 관측치들의 오차항의 가중 평균
        - 오차항 : 예측 오차를 이전 관측치들의 오차항으로 설명
        - 이동평균 차수 : 이동평균 차수는 몇 개의 이전 관측치들의 오차항을 사용 현재 시점의 값 예측
        - 자기 상관성 : 이전 관측치들의 오차항들은 서로 상관되지 않는다고 가정

    <br>
