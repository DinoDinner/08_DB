#### 트랜드

1. 데이터가 증감하는 패턴
2. 데이터의 추세를 보이며, 데이터의 특성을 파악하는 데 도움

- 상향(Upward) : 긍정적인 요인의 의해 발생
- 하향(DownWard) : 부정적인 요인에 의해 발생
- 수평(Flat) : 안정적인 상태나 균형

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

#### 1. 데이터 생성

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range (start='2023-01-01', periods=100, freq='D')
values = [10,12,11,15,9,13,8,14,10,16] * 10 # 시계열 데이터

# 데이터프레임 생성
data = pd.DataFrame({'Date': dates, 'Value': values})
```

#### 2. 주기성 설정 및 분해, 추세 계산

```python
# 주기 설정
period = 7  # 7일 주기

# seasonal_decompose 모델 사용해서 분해
result = seasonal_decompose(data['Value'],model='addictive',period=period)

# 추세 계산
trend = result.trend # 추세 가져오기
print(trend)
```

#### 3. 시각화

```python
#시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], label='Original')
plt.plot(data['Date'],trend, label='Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=-45)
plt.title('Trend Extraction In Time Series Data')
plt.legend()
plt.show()
```

- 파란색 : 원본 시계열 데이터
- 주황색 : 7일마다 추세를 분석한 시계열 데이터
- 7일마다 상승하는 추세경향을 보임
