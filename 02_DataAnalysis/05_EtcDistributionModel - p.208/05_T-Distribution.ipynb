#### T-분포

1. 통계적 가설 검정에 활용됨.
2. 두 그룹 간의 평균 차이를 비교하거나 회귀분석에서 회귀 계수의 유의성을 평가
3. 표본 평균의 신뢰구간을 추정할 수 있음. 신뢰구간은 표본크기와 자유도에 따라 달라짐
4. 정규 분포 가정이 충족되지 않는 작은 표본 크기에서 유용하게 활용될 수 있음. 또는 모집단의 분산이 알려져 있지 않은 경우

- 자유도에 따라 모양이 변화, 커질 수록 표준 정규 분포에 근사함.
- 표본 크기가 작을 때 표본 평군의 추정이 불확실할 수 있음
- 표본크기와 자유도에 따라 CDF가 달라지며, 주어진 값보다 작은 확률을 계산할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = 10 # 자유도
num_simulation = 1000
t_values = np.random.standard_t(df,num_simulation)

# 시각화
plt.hist(t_values, bins=20, alpha=0.5)
plt.show()
```

- T 분포의 모양은 자유도에 따라 달라짐
- T 분포의 평균이 0을 중심으로 대칭적
