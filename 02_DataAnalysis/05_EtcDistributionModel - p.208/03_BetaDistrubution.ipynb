#### 베타분포

1. 0-1 사이의 값을 가지는 확률 변수를 모델링 하는데 사용되는 연속형 확률분포
2. alpha/beta를 모수를 가지며, 이에 따라 분포의 형태가 결정

- 범위가 제한 되어 있고 유연성을 가져 다양한 모양의 분포를 형성 할 수 있음
- 베이지안 추론 / 베타-이진 모델링 / 확률계산을 모델링 등에 사용됨
- CDF/PDF를 사용하여 확률을 계산함.

- A/B 테스트 : 두 가지 버젼의 그룹을 만들어서 각 그룹의 성과 확률을 모델링 (상품 비교)
- 성공확률 모델링 : 이항분포의 성공확률 모델링하는데 사용 (제품 결함, 고객 구매 확률)
- 베이지안 추론 : 베타분포를 사전분포로 사용하여 데이터 관찰 후 사후 분포를 계산하고 추로한다. (농업)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
# 모수 설정
alpha = 2
beta = 5

# 랜덤으로 사용자 행동 데이터 생성(0-1 사이의 값)
user_data = np.random.beta(alpha,beta,size=1000)
print(user_data)
```

```python
# 데이터 분포 시각화
plt.hist(user_data, bins=30, density=True, alpha=0.7)
plt.xlabel('User Behavior')
plt.ylabel('Probality Density')
plt.title('User Behavior Distribution (alpha={}, beta={})'.format(alpha,beta))
plt.show()
```

- alpha, beta 값의 따라 모양이 결정 됨
- 베타분포 모형에 따라 사용자 행동 데이터의 분포를 시각화 할 수 있음
