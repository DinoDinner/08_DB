#### 예시 데이터에 적합한 정규화와 변환 방법

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

#### 1. 비정규 분포 형태 데이터 생성

```python
# 예시 데이터 생성
data = pd.DataFrame({'Value':[10, 20, 30, 40, 50]})

# 데이터 분포 확인
plt.hist(data['Value'])
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Data Distribution')
plt.show()
```

#### 2. 로그변환

- 로그 변환을 적용한 후의 데이터 분포가 이전과 비교하여 더 정규분포에 가까워진 것을 확인
- 로그 변환을 통해 더 넗은 범위로 퍼져 있는 것을 알 수 있음

```python
# 로그 변환 적용
data['LogValue'] = np.log(data['Value'])

# 변환된 데이터 분포 확인
plt.hist(data['LogValue'])
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Log Transformed Data Distribution')
plt.show()
```

#### 3. 박스-콕스 변환

- 박스-콕스 변환을 적용한 후의 데이터 분포가 이전과 비교하여 더 정규분포에 가까워진 것을 확인
- 변환된 데이터는 더 넓은 범위로 퍼져 있으며 데이터가 정규 분포의 형태를 띠고 있음을 알 수 있음

```python
# 박스-콕스 변환 수행
transformed_data, lambda_ = boxcox(data['Value'])

# 변환된 데이터 분포 확인
plt.hist(transformed_data)
plt.xlabel('Transformed Data')
plt.ylabel('Frequency')
plt.title('Box-Cox Transformed Data Distribution')
plt.show()

# 변환된 데이터와 lambda 값 출력
print("Transformed Data:")
print(transformed_data)
print("Lambda value:", lambda_)
```
