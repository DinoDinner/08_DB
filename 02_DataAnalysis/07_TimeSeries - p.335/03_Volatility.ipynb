#### 시계열 데이터의 변동성

데이터 값을 사이의 변동의 정도나 크기를 의미한다.

- 측정지표
    1. 분산 : 데이터 값들의 퍼짐도를 측정하는 지표
    2. 표준편차 : 데이터 값들의 퍼짐 정도를 나타내는 지표
    3. 변동계수 : 상대적인 변동의 정도를 나타내는 지표

- 변동성은 시계열 데이터의 불안정성을 파악하는데 중요한 지표로 활용
- 변동성을 이해하고 적절히 처리하는 것이 중요함.

#### 시계열 데이터 시각화 기법 ( 선 그래프, 박스 플롯, 자기상관 그래프 )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

####  1. 선 그래프

```python
# 시계열 데이터 생성
np.random.seed(42)
dates_line = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values_line = np.random.randn(1000)

# 데이터 프레임 생성
data_line = pd.DataFrame({'Date':dates_line,'Value':values_line})

# 시계열 데이터의 선 그래프
plt.plot(data_line['Date'],data_line['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Line Plot')
plt.xticks(rotation=-45)
plt.show()
```

#### 2. 산점도 그래프

```python
# 시계열 데이터 생성
np.random.seed(42)
dates_sct = pd.date_range(start='2023-01-11', periods=1000, freq='D')
valeus_sct = np.random.randn(1000)

# 데이터 프레임 생성
data_sct = pd.DataFrame({'Date':dates_sct,'Value':valeus_sct})

# 시계열 데이터의 산점도 그래프
plt.scatter(data_sct['Date'],data_sct['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=-45)
plt.title('Time Series Scatter Plot')
plt.show()
```

#### 3. 막대 그래프 시각화

```python
# 시계열 데이터 생성
np.random.seed(42)
dates_bar = pd.date_range(start='2023-01-11', periods=1000, freq='D')
valeus_bar = np.random.randn(1000)

# 데이터 프레임 생성
data_bar = pd.DataFrame({'Date':dates_bar,'Value':valeus_bar})

# 시계열 데이터의 막대 그래프
plt.bar(data_bar['Date'],data_bar['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=-45)
plt.title('Time Series Bar Plot')
plt.show()
```

#### 4. 박스 플롯 시각화

```python
# 시계열 데이터 생성
np.random.seed(42)
dates_box = pd.date_range(start='2023-01-11', periods=1000, freq='D')
valeus_box = np.random.randn(1000)

# 데이터 프레임 생성
data_box = pd.DataFrame({'Date':dates_box,'Value':valeus_box})

# 시계열 데이터의 박스플롯 그래프
plt.boxplot(data_box['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Box Plot')
plt.show()
```

#### 5. 히스토그램 분포 시각화 

```python
# 시계열 데이터 생성
np.random.seed(42)
dates_hist = pd.date_range(start='2023-01-11', periods=1000, freq='D')
valeus_hist = np.random.randn(1000)

# 데이터 프레임 생성
data_hist = pd.DataFrame({'Date':dates_hist,'Value':valeus_hist})

# 시계열 데이터의 히스토그램 그래프
plt.hist(data_hist['Value'],bins=100)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Histogram')
plt.show()
```

#### 6. 히트맵 그래프 시각화

```python
import seaborn as sns

# 시계열 데이터 생성
np.random.seed(42)
dates_hm = pd.date_range(start='2023-01-11', periods=1000, freq='D')
valeus_hm = np.random.randn(1000)

# 데이터 프레임 생성
data_hm = pd.DataFrame({'Date':dates_hm,'Value':valeus_hm})

# 데이터를 피벗 테이블로 변환 (2차원 형태)
date_hm_pivot = data_hm.pivot(index='Date',columns='Value',values='Value')

# 시계열 데이터의 히트맵 그래프
plt.figure(figsize=(10,6))
sns.heatmap(date_hm_pivot, cmap='ocean',cbar=True)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Heatmap')
plt.show()
```


- 피벗 테이블 : 데이터를 특정 열을 기준으로 정리하는 것.
