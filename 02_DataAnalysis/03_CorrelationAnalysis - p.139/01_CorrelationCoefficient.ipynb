#### 상관 분석

1. 두 변수 간의 관계를 살펴보기 위한 방법 중 하나.
2. 1에 가깝다면 양, -1에 가깝다면 음의 상관관계를 가진다.

```python
import seaborn as sns
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
print(tips)
```

```python
corr = tips[['total_bill','tip']].corr()
print(corr)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
# calculate Pearson correlation coefficient between total_bill and tip
corr = tips['total_bill'].corr(tips['tip'], method='pearson')

print('Pearson correlation coefficient: ',corr)
```

- tip와 total bill의 상관계수가 0.6이다 = 상관이 있다.
