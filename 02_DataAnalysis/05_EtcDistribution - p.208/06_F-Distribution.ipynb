#### F-분포

1. 분산 비교 및 분산 분석에서 사용되는 확률 분포.
2. 두 개 이상의 그룹의 분산이 동일한지를 비교하거나, 회귀 분석의 모델 적합도를 평가하는데 사용
3. 두 개의 자유도 df1/df2로 정의
4. df1 : 분자의 자유도, 분산비교에 참여하는 그룹 수
5. df2 : 분모의 자유도, 분산을 계산하는 표본의 개수에서 분자의 자유도를 뺀 값

- 대칭성을 가지지 않고 비대칭적
- 항상 양수의 값을 가지고, 정의 범위는 0에서 무한대
- 자유도(df1,df2)에 따라 분포현태 변화
- 적절한 가정과 조건을 충족해야 올바른 결과를 얻을 수 있음
- 확률 계산은 CDF를 사용하여 이루어지고, df1과 df2에 따라 정의 됨.
- 분산비교, 회귀분석 모델의 적합도 평가, 분산 분석, 자유도 검정에 활용됨

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# F-분포 모델링
dfn = 5 # 자유도 1
dfd = 10 # 자유도 2

# F-분포 시뮬레이션
num_simulation = 1000 # 시뮬레이션 횟수
f_values = np.random.f(dfn, dfd, num_simulation)
print(f_values)
```

```python
# F-분포 시뮬레이션 결과 시각화
plt.hist(f_values, bins=20, alpha=0.7)
plt.xlabel('F-Value')
plt.ylabel('Frequency')
plt.title('F-Distribution Simulation')
plt.show()
```
