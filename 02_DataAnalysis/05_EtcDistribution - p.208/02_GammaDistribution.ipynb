#### 감마 분포

1. 연속 확률분포의 하나로, 양의 실수 값을 가지는 변수에 대해 사용
2. 특정 사건이 발생하기까지 걸리는 대기시간, 제품의 수명, 신뢰성 등을 모델링 하는데 사용
3. 지수분포는 감마분포의 특수한 경우로 볼 수 있다.

- 양수인 실수 값
- shape parameter(형상 모수), rate parameter(비율 모수)를 가짐
- 형상모수 : 분포의 모양 / 비율 모수 : 분포의 폭 결정
- CDF, PDF를 이용하고, 두 개의 모수를 조절해서 분포의 모형 조절

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
# 제품의 고장 까지 걸리는 시간 모델링 하는 감마분포 생성함수
def generate_lifetime(alpha, beta, size) :
    return np.random.gamma(shape=alpha, scale=beta, size=size)

def measure_lifetime(alpha, beta, num_samples) :
    lifetime = generate_lifetime(alpha, beta, num_samples)
    
    # 고장까지 걸리는 시간 분포 히스토그램
    plt.hist(lifetime, bins=30, density=True, alpha=0.8) 
    plt.xlabel('Lifetime')
    plt.ylabel('Provbaility Density')
    plt.title('Product Lifetime Distribution')
    plt.show()
    
# 고장까지 걸리는 시간 분포 측정 및 시각화  
measure_lifetime(2.5, 1.2, 1000)

# 모수 설정
# alpha = 2.5 # 감마분포의 형상 모수
# beta = 1.2 # 감마분포의 비율 모수
# num_samples = 1000 # 생성할 샘플 수
```

- alpha가 클수록 형태는 좌측으로 왜곡된다.
- beta가 클수록 높이는 낮아지고, 폭은 넓어진다.

- 제품의 고장까지 걸리는 시간 분포를 시각적으로 확인할 수 있음.
- 분포의 모양과 높이를 통해 제품의 고장까지 걸리는 평균 시간과 분산 등의 특성을 파악할 수 있다.
