#### 지수분포

1. 사건이 일정 간격 또는 시간 내에 발생하는데 사용되는 연속 확률 분포
2. 사건 발생 간격이 지수적으로 분포한다 가정
3. 사건 간격은 항상 양수, 지수 분포는 0이상의 실수
4. 오래된 사건 일수록 발생 가능성이 낮아짐

- PDF(확률 밀도 함수), CDF(누적 분포 함수), 평균과 분산을 이용해서 계산
- 과거 이벤트에 대한 정보가 미래 사건에 영향을 주지 않은 Memoryles(무기억성) 특성을 가진다. = 과거 사건 발생해도 앞으로의 확률은 변동이 없음
- 사건 간격 대기 시간, 서비스 대기 시간 등 연속적인 사건 발생을 모델링하고 예측하는데 활용

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 고객 서비스 센터 대기 시간을 모델링하는 지수 분포 생성함수
def generate_wating_time(lam,size) :
    data =  np.random.exponential(scale=1/lam, size=size)
    return data

# 고객 서비스 센터 대기 시간 분포 측정 및 시각화
def measure_wating_time(lam,num_samples) :
    wating_times = generate_wating_time(lam, num_samples)
    
    # 대기 시간 히스토그램
    plt.hist(wating_times, bins=30, density=True, alpha=0.7)
    plt.xlabel('Wating Time')
    plt.ylabel('Probalitiy Density')
    plt.title('Customer Service Center Waiting Time Distribution')
    plt.show()

# 모수 설정
lam = 0.5 # 지수분포의 lambda 값

# 모수 설정
num_samples = 1000  # 생성할 샘플 수
measure_wating_time(lam,num_samples)
```

- lam = 0.5 로 설정했기 때문에, 평균 대기시간은 1/lam = 2시간이다.
- 대기 시간 분포의 평균은 약 2시간이 될 것임을 예상 가능.
- 대부분의 고객들이 0-4까지 짧은 대기 시간을 경험한다는 것을 시각적으로 확인할 수 있다.
