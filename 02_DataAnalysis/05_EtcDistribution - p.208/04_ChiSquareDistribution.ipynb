#### 카이제곱 분포

1. 정규분포를 따르는 확률변수들을 제곱하여 합산
2. 분산 추정, 독립성 검정, 적합도 검정 등에 사용
3. 분포의 형태는 자유도에 의해 결정
4. 카이제곱 분포는 비음수 , 평균은 자유도에 의존, 분산은 2배의 자유도

- 자유도가 커질 수록 분포의 형태는 정규분포에 가까워지고, 분산이 커진다.
- 카이제곱 분포의 확률 계산은 CDF를 사용해서 수행.
- 확률계산을 통해 카이제곱 분포의 특정 값레 대한 확률을 구할 수 있고, 통계적 추론이나 가설 검정 등에 활용
- 독립성 검정, 클러스터 간의 분산평가, 통계적 모델의 적합성 평가, 신뢰도 분석, 자료의 적합성 검정에 활용된다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 모수 설정
df = 5 # 자유도 (degrees of freedom)
```

```python
# 카이제곱 분포에서 랜덤 변수 생성
random_variable = np.random.chisquare(df,size=1000)

# 분포 시각화
plt.hist(random_variable, bins=30, density=True, alpha=0.7)
plt.xlabel('Random Varibale')
plt.ylabel('Probabilty Density')
plt.title('Chi-Squared Distribtuion (df={})'.format(df))
```

- 자유도가 증가할수록 분포의 모양이 좀 더 대칭에 가까워짐.
- 자유도가 증가할수록 분포의 분산이 감소
