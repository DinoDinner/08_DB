#### 이항분포

1. 독립적인 사건이 두 가지 결과만을 가지는 경우 사용되는 확률분포
2. 두 가지 결과 중 하나의 결과를 갖는 여러 번의 독립적인 시행에서 발생하는 확률 현상을 모델링 할때 유용하게 사용

- 이항분포의 확률변수 : 성공횟수 / 정수값
- 성공 : 고정된 확률(p)로 가정 / 실패 : 1-p
- 시행 횟수 : n / 성공 횟수 : k

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# 이항분포 모델링
n = 10 # 시행횟수
p = 0.5 # 성공확률
```

```python
# 이항분포 시뮬레이션
num_simulation = 1000 # 시뮬레이션 횟수
successes = np.random.binomial(n,p,num_simulation)
```

```python
# 이항분포 시뮬레이션 결과 시각롸
plt.hist(successes, bins=11,range=(0,10), alpha=0.5)
plt.xlabel('Successes')
plt.ylabel('Frequency')
plt.title('Binormal Distrubution Simulation')
plt.show()
```
