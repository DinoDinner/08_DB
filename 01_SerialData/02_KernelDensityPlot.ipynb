#### 밀도 그래프
밀도 그래프(Kernel Density Plot)는 자료의 분포 상태를 부드러운 곡선으로 나타낸 그래프입니다.

예를 들어, 0에서 100까지의 점수를 가진 학생들의 점수 분포를 밀도 그래프로 나타내면, 각 점수값을
따라 부드러운 곡선이 그려질 것입니다. 

이때 곡선의 높낮이는 해당 구간에서 점수값이 나올 확률(밀도)
을 나타내게 됩니다.

밀도 그래프는 히스토그램과 비교하여 부드러운 곡선으로 자료의 분포를 더 자세히 파악할 수 있습니다.

또한, 히스토그램보다는 데이터의 분포 형태를 쉽게 파악할 수 있어서 데이터 분석에서 널리 사용됩니다.

예를 들어, 어떤 상품의 판매량을 밀도 그래프로 나타내면, 상품의 인기도, 수요 분포 등을 더욱 자세히
파악할 수 있습니다.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터셋 불러오기
data = pd.read_csv('../99_Data/02_Data/BlackFriday.csv')
print(data,'\n')

# User_ID별 구매 총 금액 계산
user_total_spent = data.groupby('User_ID')['Purchase'].sum()
print(user_total_spent)
```

```python
# 밀도 그래프 시각화

plt.figure()
user_total_spent.plot.kde()
# Kernel Desnsity Estimate(커널 밀도 추정) : 데이터의 밀도를 보여주는 그래프

plt.title('Density Plot of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')

plt.show()
```

-  0 - 0.2 사이에 빈도수가 가장 높다.
- 선의 높이가 높을수록 해당 범위의 데이터가 더 많이 분포해 있다.
- 높은 부분은 해당 데이터값이 많이 분포하는 구간을 나타내며, 낮은 부분은 그 반대이다.
