# 밀도 그래프
---

밀도 그래프(Kernel Density Plot)는 자료의 분포 상태를 부드러운 곡선으로 나타낸 그래프입니다.

예를 들어, 0에서 100까지의 점수를 가진 학생들의 점수 분포를 밀도 그래프로 나타내면, 각 점수값을
따라 부드러운 곡선이 그려질 것입니다. 

이때 곡선의 높낮이는 해당 구간에서 점수값이 나올 확률(밀도)
을 나타내게 됩니다.

밀도 그래프는 히스토그램과 비교하여 부드러운 곡선으로 자료의 분포를 더 자세히 파악할 수 있습니다.

또한, 히스토그램보다는 데이터의 분포 형태를 쉽게 파악할 수 있어서 데이터 분석에서 널리 사용됩니다.

예를 들어, 어떤 상품의 판매량을 밀도 그래프로 나타내면, 상품의 인기도, 수요 분포 등을 더욱 자세히
파악할 수 있습니다.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#  정규분포를 따르는 데이터 생성

data = np.random.normal(loc=0, scale=1, size=10000)

# 데이터 분포 시각화

sns.histplot(data)
plt.show()

# 데이터 통계 지표 계산
mean = np.mean(data)
std = np.std(data)
print("평균 : " , mean)
print("표준편차 : ", std)
```
