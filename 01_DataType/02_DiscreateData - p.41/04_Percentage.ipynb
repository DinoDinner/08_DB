#### 백분율

1. 상대 빈도수를 100으로 곱한 값
2. 비율을 %로 표현 할 수 있다.

- 100을 기분으로 어떤 값이 차지하는 비율을 나타냄

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 만들기
labels = ['A','B','C','D','E']

# 데이터 비율 만들기
percentages = [30,20,15,10,25]
```

```python
# 원형 그래프

fig, ax = plt.subplots()
ax.pie(percentages, labels=labels, autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Percentages of Categories')
plt.show()
```

```python
# 막대 그래프

fig,ax = plt.subplots()
ax.bar(labels,percentages)

ax.set_xlabel('Labels')
ax.set_ylabel('Percentages')
ax.set_title('Percentages of Labels')

plt.show()
```
