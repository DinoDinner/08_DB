#### 이산형 데이터

1. 정수 값으로 이루어진 데이터
2. 연속적이지 않고 한정된 개수의 값으로 이루어짐
3. 예시 : 동전 던지기 앞/뒤 개수

- 이산형 데이터는 이산형 변수의 값으로 사용.
- 연속형 데이터와 달리 측정 값의 범위가 한정되어 있음.
- Histogram로 시각화 하고, 대표값으로 Mode 사용.
- 대표값 선정이 중요하고, 범주형 데이터와 함께 분석하면 더 의미있는 결과를 도출 할 수 있음.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
data = pd.read_csv('../00_DataEx/02_College_data.csv')

print(data['top10perc'].unique()) # type: ignore
# data 중 top10perc column 불러오기
# top10perc column은 대학교의 학생 중 상위 10% 학생의 비율을 나타내는 이산형 데이터
# 이 Column을 사용하여 이산형 데이터 시각화를 진행
# .unique : index를 없애준다.

# 이산형 데이터라 모두 정수로 나온다.
```

```python
print(data)

# top10perc에서 private column yes = private / no = public
```

```python
# top10perc 막대 그래프 그리기

plt.figure(figsize=(10,6))   # figure : 그래프의 옵션 설정 (너비,높이)
plt.bar(data['private'],data['top10perc'])  # bar : 막대그래프

plt.title('Bar Chart of Top 10% Scores by Private/Public')
plt.xlabel('Private/Public')
plt.ylabel('Top 10% Scores')

plt.show()
```

- 1번 그래프 : Private 학교의 top10perc
- 2번 그래프 : Public 학교의 top10perc

- Private의 성적이 좀 더 높다.

```python
# top10perc 원 그래프 그리기

plt.figure(figsize=(10,6))
data.groupby('private')['top10perc'].sum().plot(kind='pie',autopct='%1.1f%%')
# private와 top10perc의 데이터를 그룹으로 지정한다.
# Sum은 Private의 top10perc의 데이터를 모두 합한다.
# %원하는 양식%으로 하면 (%)도 함께 출력된다.

plt.title('Pie Chart of Top 10% Scores by Private/Public')
plt.ylabel('')
# '' : Private Yes/No를 자동으로 보여줌

plt.show()
```

- Private의 top10perc이 전체 top10perc의 77.4%를 차지하고 있다.

```python
# top10perc 히스토그램 그리기

plt.figure(figsize=(10,6))
plt.hist(data['top10perc'],bins=20)

plt.title('Histogram of Top 10% Scores')
plt.xlabel('Top 10% Scores')
plt.ylabel('Frequency')

plt.show()
```

- top10perc의 점수 분포를 히스토그램으로 보여줌
- 10 ~ 30 점수 구간에서 top10perc이 가장 높은 빈도를 보여줌.
