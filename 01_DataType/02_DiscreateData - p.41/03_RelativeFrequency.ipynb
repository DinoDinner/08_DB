#### 상대 빈도수

1. 빈도수/전체 데이터 수
2. 각 값이 전체에서 차지하는 비율을 알 수 있다.

- 빈도수와 달리 전체 데이터의 크기에 상관없이 각 데이터 값의 비율을 비교할 수 있음
- 각 데이터의 특성을 파악하는 데 유용하다
- 0과 1 사이의 값이기에, 백분율로 변환해서 해석하는게 편하다


----
#### 계산 방법
- 각 숫자 빈도수 체크
- 만약 예를 들어 5라는 숫자의 빈도수는 4번
- 4/11(데이터 총 수) = 0.36363636


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 예시 데이터 생성 1
data = np.array([1,2,3,3,4,4,4,5,5,5,5])

# 데이터에서 각 값의 빈도수 계산
value_counts = np.bincount(data)    # 0:0개 , 1:1개, 2:2개 ...
print(value_counts)

# 각 값의 상대빈도수 계산
relative_frequencies = value_counts/len(data)
print(len(data))    # 데이터의 개수
print(relative_frequencies) # 빈도수 계산
```

```python
# 예시 데이터 생성 2
from collections import Counter
 
def print_frequency(numbers):
    # 빈도수 계산
    counts = Counter(numbers)
 
    # (빈도수, 수) 쌍의 리스트 생성
    frequency_list = [(count, number) for number, count in counts.items()]
 
    # 빈도수를 기준으로 정렬
    sorted_list = sorted(frequency_list)
 
    # 결과 출력
    for count, number in sorted_list:
        print(f"수: {number}, 빈도수: {count}")
 
# 예시 입력 데이터
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
 
# 함수 호출
print_frequency(data)
```

```python
# 막대 그래프로 상대빈도수 시각화

plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(np.arange(len(value_counts)), np.arange(len(value_counts))+1)

plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Bar Chart')

plt.show()
```
