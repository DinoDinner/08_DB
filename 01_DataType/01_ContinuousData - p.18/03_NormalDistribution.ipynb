#### 정규 분포

- 정규분포는 통계학에서 가장 중요하고 많이 사용되는 분포 중 하나
- 그래프는 종모양으로 나타나며, 대부분의 자연현상에서 나타나는 데이터 분포와 비슷하다.
- 예시 : 인구의 키나 체중, 공장에서 생산되는 제품의 길이나 무게, 시험성적 등

```python
import numpy as np
import seaborn as sns   # seaborn : 그래프를 그리는 메소드
import matplotlib.pyplot as plt
```

```python
#  정규분포를 따르는 데이터 생성

data = np.random.normal(loc=0, scale=1, size=10000)
# 평균 : 0  / 분산 : 1 인 정규분포 데이터
print(data)
# 데이터 분포 시각화
# list type 데이터로 나옴

sns.histplot(data)
plt.show()

# 데이터 통계 지표 계산
mean = np.mean(data)
std = np.std(data)
print("평균 : " , mean)
print("표준편차 : ", std)
```

- 평균을 중심으로 좌우대칭형의 분포 = 정규분포를 따른다.
- 평균과 표준편차가 매우 중요한 역할을 한다.
- 데이터 분석에 필수적인 정보이다.
