#### 상관관계

- 연속형 데이터의 상관관계는 두 변수 사이의 관계를 파악할 떄 유용한 지표다.
- Correaltion Coefficient(상관계수)를 사용하여 파악

1. 두 변수 간의 어떤 관련성이 있는지 나타내는 지표
2. 함께 변화하는 경향 and 관계가 얼마나 강한지 파악할 수 있음
3. 1에 가까울 수록 양의 상관관계 / -1에 가까울 수록 음의 상관관계 / 0은 관계 없음.

```python
# 데이터 생성
np.random.seed(123)
x=np.random.normal(loc=0, scale=1,size=100)
y=2*x+np.random.normal(loc=0,scale=1,size=100)

# 선점도 시각화
fig, ax = plt.subplots()
ax.scatter(x,y)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot')

plt.show()

# 상관관계 계산
corr = np.corrcoef(x,y)[0,1]
print("상관계수 : ", corr)
```

- 상관계수 : 0.9는 강한 양의 상관관계를 갖는다는 걸 알려준다.
