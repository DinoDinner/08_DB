#### 중심 경향성

1. 연속형 데이터의 중심 경향성을 파악하는 것이 중요.<br>
2. 대표적으로 평균/중앙값/최빈값으로 계산할 수 있다.<br>
3. 데이터 분포가 비대칭적이면 평균보다 중앙값이나 최빈값이 더 적합한 경우가 있음.<br>

- Mean(평균)은 데이터 값들의 총합/개수
- Median(중앙값)은 데이터의 크기를 순서대로 나열했을 때 가운데에 위치하는 값
- Mode(최빈값)은 데이터에서 가장 자주 등장하는 값


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
data = np.random.normal(loc=0,scale=1,size=10000)
# loc(평균) / scale(표준편차) / size(데이터 량)

print(data)

# 중심 경향성 계산
mean = np.mean(data)
median = np.median(data)
mode = np.round(np.mean(data))  # np.round : 반올림
```

```python
# 분포 시각화
fig, ax = plt.subplots()

# 데이터 그래프
ax.hist(data, bins=50)

# 평균/중앙값/최빈값 라인
ax.axvline(mean, color='red', linestyle='dashed', linewidth=2)
ax.axvline(median, color='green', linestyle='dashed', linewidth=2)
ax.axvline(mode, color='purple', linestyle='dashed', linewidth=2)

# 그래프 라벨 붙이기
ax.legend(['Mean: {:.2f}'.format(mean), 'Median: {:.2f}'.format(median), 'Mode: {}'.format(mode)])
ax.set_xlabel('Data Values')
ax.set_ylabel('Frequency')
ax.set_title('Data Distribution')

plt.show()

# 해당 그래프는 평균(mean), 중앙값(median), 최빈값(mode)을 구하고 이를 시각화했다.
```

- 모두 0에 근접한 값을 가진다.
- 최빈값은 loc를 0으로 설정했기 때문에 0으로 나온다.
